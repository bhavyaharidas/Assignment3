```python
from tpot import TPOTClassifier
from sklearn.model_selection import train_test_split
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("loan.csv", low_memory =False)
pd.set_option('display.max_columns',None)
df.fillna(0,inplace=True)
```

```python
# Drop unneccesary columns
df = df.drop(['url', 'desc', 'policy_code', 'last_pymnt_d', 'next_pymnt_d', 'earliest_cr_line', 'emp_title'], axis=1)
df = df.drop(['id', 'title', 'total_rec_int', 'total_rec_late_fee', 'total_rec_prncp', 'zip_code'], axis=1)
```

```python
# create an bad/good loan indicator feature
df['bad_loan'] = np.where((df.loan_status == 'Charged Off') |
                          (df.loan_status == 'Does not meet the credit policy.  Status:Charged Off') | 
                          (df.loan_status == 'Does not meet the credit policy.  Status:Current') | 
                          (df.loan_status == 'In Grace Period')|
                          (df.loan_status == 'Late (31-120 days)A')|
                          (df.loan_status == 'Default')
                          , 0, 1)

#bad loan column ==  1 then loan is bad else if  == 0 then good loan
```

```python
import pandas as pd  
from sklearn.model_selection import train_test_split  
from tpot import TPOTClassifier  
  
loan = pd.read_csv('Cleaned_loan.csv', low_memory =False)  
  
X = loan.drop('bad_loan', axis=1).values  
y = loan.loc[:, 'bad_loan'].values  
  
X_train, X_test, y_train, y_test = train_test_split(X, y,  
                                                    train_size=0.75,  
                                                    test_size=0.25)  
  
my_tpot = TPOTClassifier(generations=10)  
my_tpot.fit(X_train, y_train)  
  
print(my_tpot.score(X_test, y_test))  
```
